```python
import pandas as pd
import numpy as np
dataset=pd.read_csv('https://raw.githubusercontent.com/scalabretta/GroupProject--IMDB/master/Dataset/IMDB-database-update.csv',encoding='latin1')
dataset.head()
df1=pd.DataFrame(dataset)
```

```python
#No 1 How many rows and columns reflected on dataset file IMDB-database-update
df1.shape
```

```python
#soal no 2 top 3 movies highest net profit in 2013
df1[df1['title_year']==2013].sort_values(by='net_USD',ascending=False,).head(3)
```

```python
#soal no 3 top movies with highest gross prodit in 2009
df1[df1['title_year']==2009].sort_values(by='gross_USD',ascending=False).head(3)
```

```python
#No 4 how many non-english movies are in the dataset?
df1[df1['language']!='English'].count().head(1)
```

```python
#No 5 What are top languages in the dataset aside english
non_english=df1.where(df1.language!='English')
non_english.language.value_counts().head(3)
```

```python
#No 6 what are the top 3 countries with the highest nett profit in all dataset times
df1.groupby('country').sum().sort_values(by='net_USD',ascending=False).head(3)

```

```python
#No 7 what are the average budget of all movies in dataset in 2013
average_2013=df1[df1['title_year']==2013].budget_USD.mean()
average_2013
```

```python
#No 8 what are the average budge of all movies in dataset in 2014
average_2014=df1[df1['title_year']==2014].budget_USD.mean()
average_2014
```

```python
#No 9 Which one is higher, average budget of all movies in dataset in 2014 or 2013
if average_2013>average_2014:
    print(' average budget 2013 is more than 2014')
else:
    print(' average budget 2014 is more than 2013')
```
